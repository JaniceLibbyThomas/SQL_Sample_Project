```python
import pymongo
#to establish  connection

client = pymongo.MongoClient("mongodb+srv://jane:1234@cluster0.dytvv0d.mongodb.net/?retryWrites=true&w=majority")
db = client.StudentDetails #d30 -->database name
records=db.Marks  #apple-->colleection name

#Question 1
value=[]

test_type=[]


for i in records.aggregate([{'$unwind':'$scores'},{'$group':{'_id':'$scores.type','max':{'$max':'$scores.score'}}}]):
  value.append(i['max'])
  test_type.append(i['_id'])

d={}
ID={}
ID1={}
for j in range (0,len(value)):
  for i in records.aggregate([{'$unwind':'$scores'},{'$match': {'scores.score':value[j],'scores.type':test_type[j]}}]):
    d[(i['_id'])]= str(i['name'])
    ID[str(i['_id'])]=i['scores']['type']
    ID1[str(i['_id'])]=i['scores']['score']
import pandas as pd
df=pd.DataFrame([d],index=['MAX SCORE - ID & NAME ']).T
df['TYPE'] = ID.values()
df['SCORE'] = ID1.values()
display(df)
```

```python
#Question 2
# 2)      Find students who scored below average in the exam and pass mark is 40%?

import pymongo
#to establish  connection

client = pymongo.MongoClient("mongodb+srv://jane:1234@cluster0.dytvv0d.mongodb.net/?retryWrites=true&w=majority")
db = client.StudentDetails #d30 -->database name
records=db.Marks  #apple-->colleection name

count =0
Sum =0

test_type=[]

for i in records.aggregate([{'$unwind':'$scores'},{'$match': {'scores.score':{'$gte':40},'scores.type':'exam'}},{'$group': {'_id':0,'count':{'$sum':1}}}]):
  count=(i['count'])

for i in records.aggregate([{'$unwind':'$scores'},{'$match': {'scores.score':{'$gte':40},'scores.type':'exam'}},{'$group': {'_id':1,'Sum':{'$sum':'$scores.score'}}}]):
  Sum=(i['Sum'])

average = (Sum / count)
d={}
ID={}
for i in records.aggregate([{'$unwind':'$scores'},{'$match': {'scores.score':{'$lt':average,'$gte':40},'scores.type':'exam'}}]):
  d[(i['_id'])]= str(i['name'])
  ID[str(i['_id'])]=i['scores']['score']

import pandas as pd
df=pd.DataFrame([d],index=['ID & NAME - BELOW AVERAGE STUDENT LIST']).T
df['EXAM MARK'] = ID.values()
display(df)
```

```python
#QUESTION 3
## 3)      Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.

import pymongo
#to establish  connection

client = pymongo.MongoClient("mongodb+srv://jane:1234@cluster0.dytvv0d.mongodb.net/?retryWrites=true&w=majority")
db = client.StudentDetails #d30 -->database name
records=db.Marks  #apple-->colleection name

#Question 3
d ={}
ID1={}
ID2={}
ID3={}
RESULT={}
for i in records.aggregate([{'$unwind':'$scores'},{'$match': {'scores.score':{'$gte':40}}},{'$group': { '_id': '$_id', 'count': { '$sum': 1 },'scores':{'$push':'$scores.score'},'type':{'$push':'$scores.type'}}}]):
  if(i['count']==3):
    for j in records.find({"_id":i['_id']}):
      d[j['_id']]= str(j['name'])

      ID1[j['_id']]=i['scores'][0]
      ID2[j['_id']]=i['scores'][1]
      ID3[j['_id']]=i['scores'][2]
      RESULT[j['_id']]="PASS"

for i in records.aggregate([{'$unwind':'$scores'},{'$match': {'scores.score':{'$lt':40}}},{'$group': { '_id': '$_id', 'count': { '$sum': 1 },'scores':{'$push':'$scores.score'},'type':{'$push':'$scores.type'}}}]):
  if(i['count']==3):
    
    for j in records.find({"_id":i['_id']}):
      d[j['_id']]=str (j['name'])

      ID1[j['_id']]=i['scores'][0]
      ID2[j['_id']]=i['scores'][1]
      ID3[j['_id']]=i['scores'][2]
      RESULT[j['_id']]="FAIL"

import pandas as pd
df=pd.DataFrame([d],index=['ID & NAME']).T
df['EXAM MARK'] = ID1.values()
df['QUIZ MARK'] = ID2.values()
df['HOMEWORK MARK'] = ID3.values()
df['RESULT'] = RESULT.values()
display(df)
```

```python
# Question 5

# 5)      Create a new collection which consists of students who scored below average and above 40% in all the categories.

import pymongo

myclient = pymongo.MongoClient("mongodb+srv://jane:1234@cluster0.dytvv0d.mongodb.net/?retryWrites=true&w=majority")
mydb = myclient["StudentDetails"]
records=mydb.Marks  #apple-->colleection name
mycol = mydb["Average_marks"]
mycol.drop()

d1={}
d2={}
try:
  for i in records.aggregate([{'$unwind':'$scores'},{'$match': {'scores.score':{'$gte':40}}},{'$group': {'_id':'$scores.type','count':{'$sum':1}}}]):

    d1.update({i['_id']:i['count']})

  for i in records.aggregate([{'$unwind':'$scores'},{'$match': {'scores.score':{'$gte':40}}},{'$group': {'_id':'$scores.type','Sum':{'$sum':'$scores.score'}}}]):
    d2.update({i['_id']:i['Sum']})

  av ={'homework': d2['homework']/d1['homework'],'quiz':d2['quiz']/d1['quiz'],'exam':d2['exam']/d1['exam']}


  for i in records.aggregate([{'$unwind':'$scores'},{'$match':{ 'scores.score':{'$gt':40}}},{ '$group': { '_id': '$_id', 'count': { '$sum': 1 },'scores':{'$push':'$scores.score'},'type':{'$push':'$scores.type'}}}]):
      if(i['count']==3):
      
        if(i['scores'][0]<=av['exam'] and i['scores'][1]<=av['quiz'] and i['scores'][2]<=av['homework']):
          d1 = i
          val=len(d1['scores'])
          if(val==3):
            d2 = {'_id':d1['_id'],
      'scores':[{'score':d1['scores'][0],'type':d1['type'][0]},{'score':d1['scores'][1],'type':d1['type'][1]},{'score':d1['scores'][2],'type':d1['type'][2]}]
  }
            mycol.insert_one(d2)
except:

  print("Insertion is Unsuccessful")

else:

  print('Insertion done successfully')
```
